1) Desenvolver um algoritmo que efetue a soma de todos os números ímpares que são múltiplos de
três e que se encontram no conjunto dos números de 1 até 500.

```python
def soma_impares_multiplos_de_tres(limite):
    soma = 0
    for numero in range(1, limite + 1):
        if numero % 2 != 0 and numero % 3 == 0:
            soma += numero
    return soma

limite = 500

resultado = soma_impares_multiplos_de_tres(limite)

print(f"A soma de todos os números ímpares múltiplos de três no intervalo de 1 a {limite} é: {resultado}")
```

2) Desenvolver um algoritmo que leia a altura de 15 pessoas. Este programa deverá calcular e
mostrar :
a. A menor altura do grupo;
b. A maior altura do grupo;


```python
def ler_alturas_e_calcular_extremos(num_pessoas):
    alturas = []

    for i in range(num_pessoas):
        while True:
            try:
                altura = float(input(f"Digite a altura da pessoa {i + 1} em metros: "))
                if altura > 0:
                    alturas.append(altura)
                    break
                else:
                    print("A altura deve ser um número positivo. Tente novamente.")
            except ValueError:
                print("Entrada inválida. Por favor, insira um número válido.")


    menor_altura = min(alturas)
    maior_altura = max(alturas)

    return menor_altura, maior_altura


num_pessoas = 15


menor_altura, maior_altura = ler_alturas_e_calcular_extremos(num_pessoas)


print(f"A menor altura do grupo é: {menor_altura:.2f} metros")
print(f"A maior altura do grupo é: {maior_altura:.2f} metros")


```

3) Desenvolver um algoritmo que leia um número não determinado de valores e calcule e escreva a
média aritmética dos valores lidos, a quantidade de valores positivos, a quantidade de valores
negativos e o percentual de valores negativos e positivos.

```python
def processar_valores():
    valores = []

    print("Digite os valores. Para encerrar, digite 'fim'.")

    while True:
        entrada = input("Digite um valor: ")

        if entrada.lower() == 'fim':
            break

        try:
            valor = float(entrada)
            valores.append(valor)
        except ValueError:
            print("Entrada inválida. Por favor, insira um número ou 'fim' para encerrar.")

    if not valores:
        print("Nenhum valor foi inserido.")
        return

    quantidade_total = len(valores)
    positivos = sum(1 for v in valores if v > 0)
    negativos = sum(1 for v in valores if v < 0)
    soma_total = sum(valores)

    media = soma_total / quantidade_total
    percentual_positivos = (positivos / quantidade_total) * 100
    percentual_negativos = (negativos / quantidade_total) * 100

    print(f"\nMédia aritmética dos valores: {media:.2f}")
    print(f"Quantidade de valores positivos: {positivos}")
    print(f"Quantidade de valores negativos: {negativos}")
    print(f"Percentual de valores positivos: {percentual_positivos:.2f}%")
    print(f"Percentual de valores negativos: {percentual_negativos:.2f}%")

processar_valores()


```

4) Escrever um algoritmo que leia uma quantidade desconhecida de números e conte quantos deles
estão nos seguintes intervalos: [0-25], [26-50], [51-75] e [76-100]. A entrada de dados deve
terminar quando for lido um número negativo.

```python
def contar_numeros_por_intervalo():

    contadores = {
        "0-25": 0,
        "26-50": 0,
        "51-75": 0,
        "76-100": 0
    }

    print("Digite números para contar em intervalos. Para encerrar, insira um número negativo.")

    while True:
        try:
            numero = float(input("Digite um número: "))

            if numero < 0:
                break

            if 0 <= numero <= 25:
                contadores["0-25"] += 1
            elif 26 <= numero <= 50:
                contadores["26-50"] += 1
            elif 51 <= numero <= 75:
                contadores["51-75"] += 1
            elif 76 <= numero <= 100:
                contadores["76-100"] += 1
            else:
                print("Número fora do intervalo considerado.")

        except ValueError:
            print("Entrada inválida. Por favor, insira um número.")

    print("\nContagem de números por intervalo:")
    for intervalo, contagem in contadores.items():
        print(f"Intervalo {intervalo}: {contagem} números")

contar_numeros_por_intervalo()


```

5) Faça um algoritmo estruturado que leia uma quantidade não determinada de números positivos.
Calcule a quantidade de números pares e ímpares, a média de valores pares e a média geral dos
números lidos. O número que encerrará a leitura será zero.

```python
def processar_numeros():
    quantidade_pares = 0
    quantidade_impares = 0
    soma_pares = 0
    soma_total = 0
    quantidade_total = 0

    print("Digite números positivos. Para encerrar, insira o número 0.")

    while True:
        try:
            numero = int(input("Digite um número: "))

            if numero == 0:
                break

            if numero < 0:
                print("Número negativo não é permitido. Tente novamente.")
                continue

            if numero % 2 == 0:
                quantidade_pares += 1
                soma_pares += numero
            else:
                quantidade_impares += 1

            soma_total += numero
            quantidade_total += 1

        except ValueError:
            print("Entrada inválida. Por favor, insira um número inteiro.")

    if quantidade_pares > 0:
        media_pares = soma_pares / quantidade_pares
    else:
        media_pares = 0

    if quantidade_total > 0:
        media_geral = soma_total / quantidade_total
    else:
        media_geral = 0


    print(f"\nQuantidade de números pares: {quantidade_pares}")
    print(f"Quantidade de números ímpares: {quantidade_impares}")
    print(f"Média dos valores pares: {media_pares:.2f}")
    print(f"Média geral dos números lidos: {media_geral:.2f}")


processar_numeros()


```

6) Escrever um algoritmo que gera e escreve os números ímpares entre 100 e 200.

```python
def gerar_numeros_impares(inicio, fim):
    print(f"Números ímpares entre {inicio} e {fim}:")
    for numero in range(inicio, fim + 1):
        if numero % 2 != 0:
            print(numero)

inicio = 100
fim = 200

gerar_numeros_impares(inicio, fim)


```

7) Escrever um algoritmo que leia um valor para uma variável N de 1 a 10 e calcule a tabuada de
N. Mostre a tabuada na forma: 0 x N = 0, 1 x N = 1N, 2 x N = 2N, ..., 10 x N = 10N.

```python
def calcular_tabuada(n):
    print(f"Tabuada do {n}:")
    for i in range(11):
        resultado = i * n
        print(f"{i} x {n} = {resultado}")

def main():
    while True:
        try:
            n = int(input("Digite um valor para N (de 1 a 10): "))
            if 1 <= n <= 10:
                calcular_tabuada(n)
                break
            else:
                print("Valor inválido. O valor de N deve estar entre 1 e 10.")
        except ValueError:
            print("Entrada inválida. Por favor, insira um número inteiro.")


main()



```

8) Escreva um algoritmo que leia um valor inicial A e uma razão R e imprima uma seqüência em
P.A. contendo 10 valores.

```python
def imprimir_p_a(valor_inicial, razao):
    for i in range(10):
        termo = valor_inicial + i * razao
        print(termo, end=" ")
    print()

def main():
    while True:
        try:
            a = float(input("Digite o valor inicial A: "))
            r = float(input("Digite a razão R: "))
            imprimir_p_a(a, r)
            break
        except ValueError:
            print("Entrada inválida. Por favor, insira valores numéricos.")

main()



```

9) Escreva um algoritmo que leia um valor inicial A e uma razão R e imprima uma seqüência em
P.G. contendo 10 valores.

```python
def imprimir_p_g(valor_inicial, razao):
    for i in range(10):
        termo = valor_inicial * (razao ** i)
        print(termo, end=" ")
    print()

def main():
    while True:
        try:
            a = float(input("Digite o valor inicial A: "))
            r = float(input("Digite a razão R: "))
            imprimir_p_g(a, r)
            break
        except ValueError:
            print("Entrada inválida. Por favor, insira valores numéricos.")

main()



```

10) Escreva um algoritmo que leia um valor inicial A e imprima a seqüência de valores do cálculo de
A! e o seu resultado. Ex: 5! = 5 X 4 X 3 X 2 X 1 = 120

```python
def calcular_fatorial(n):
    resultado = 1
    sequencia = []
    for i in range(n, 0, -1):
        resultado *= i
        sequencia.append(str(i))
    return sequencia, resultado

def main():
    while True:
        try:
            a = int(input("Digite um valor inicial A: "))
            if a < 0:
                print("O valor deve ser um número inteiro não negativo.")
                continue
            sequencia, resultado = calcular_fatorial(a)
            print(f"{a}! = {' X '.join(sequencia)} = {resultado}")
            break
        except ValueError:
            print("Entrada inválida. Por favor, insira um número inteiro.")

main()




```
